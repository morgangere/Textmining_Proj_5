```python
#converting to lists to transform to numeric and create new data frame
content=df['content'].to_list()
gold_standard=df['gold_standard'].to_list()
coderA=df['coderA'].to_list()
coderB=df['coderB'].to_list()
coderC=df['coderC'].to_list()
coderD=df['coderD'].to_list()
coderE=df['coderE'].to_list()
mcoderA=df['mcoderA'].to_list()
mcoderB=df['mcoderB'].to_list()
mcoderC=df['mcoderC'].to_list()
mcoderD=df['mcoderD'].to_list()
mcoderE=df['mcoderE'].to_list()
```

```python
#function to transfrom the values to numeric representations.
def replacesent(l):
    n = []
    for item in l:
        if item == 'VN':
            item = 0
        elif item == 'SN':
            item = 1
        elif item == 'N':
            item = 2
        elif item == 'SP':
            item = 3
        elif item == 'VP':
            item = 4
        n.append(item)
    return n
```

```python
#using the funtion to transfrom the data
gold_standard=replacesent(gold_standard)
coderA=replacesent(coderA)
coderB=replacesent(coderB)
coderC=replacesent(coderC)
coderD=replacesent(coderD)
coderE=replacesent(coderE)
mcoderA=replacesent(mcoderA)
mcoderB=replacesent(mcoderB)
mcoderC=replacesent(mcoderC)
mcoderD=replacesent(mcoderD)
mcoderE=replacesent(mcoderE)
```

```python
#creating an empty data frame
kdf = pd.DataFrame()
#adding each transformed list as a column
kdf['tweet']=content
kdf['gold_standard']=gold_standard
kdf['coderA']=coderA
kdf['coderB']=coderB
kdf['coderC']=coderC
kdf['coderD']=coderD
kdf['coderE']=coderE
kdf['mcoderA']=mcoderA
kdf['mcoderB']=mcoderB
kdf['mcoderC']=mcoderC
kdf['mcoderD']=mcoderD
kdf['mcoderE']=mcoderE
```

Kendall's tau-B values:
Less than + or - 0.10: very weak
+ or -0.10 to 0.19: weak
+ or - 0.20 to 0.29: moderate
+ or - 0.30 or above: strong

```python
from scipy import stats

#function to get the tau score only
def tscore(a,b):
    tau , p_value = stats.kendalltau(a,b)
    return tau
```

```python
#calculating and displaying kendell's tau for master coders
tmamb=tscore(mcoderA,mcoderB)
tmamc=tscore(mcoderA,mcoderC)
tmamd=tscore(mcoderA,mcoderD)
tmame=tscore(mcoderA,mcoderE)
tmbmc=tscore(mcoderB,mcoderC)
tmbmd=tscore(mcoderB,mcoderD)
tmbme=tscore(mcoderB,mcoderE)
tmcmd=tscore(mcoderC,mcoderD)
tmcme=tscore(mcoderC,mcoderE)
tmdme=tscore(mcoderD,mcoderE)

#calculating and displaying overall tau score of master coders
AMT_m_overall = np.mean([tmamb,tmamc,tmamd,tmame,tmbmc,tmbmd,tmbme,tmcmd,tmcme,tmdme])
AMT_m_overall
```
